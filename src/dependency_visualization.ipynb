```python
with open(r"C:\Users\Joseph Dania\Desktop\Ai_search_engine\doc\function_graph.json", "r") as file:
    source = json.load(file)
```

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.DiGraph()

for caller, callees in source.items():
    for callee in callees:
        G.add_edge(caller, callee)


pos = nx.spring_layout(
    G,
    k=3.0,        
    iterations=200,
    seed=42
)


plt.figure(figsize=(22, 22))


nx.draw_networkx_nodes(
    G,
    pos,
    node_size=2200,
    node_color="lightblue",
    edgecolors="black"
)


nx.draw_networkx_edges(
    G,
    pos,
    arrowstyle="->",
    arrowsize=20,
    width=1.6,
    alpha=0.7
)


nx.draw_networkx_labels(
    G,
    pos,
    font_size=9,
    font_weight="bold"
)

plt.title("Function Call Graph", fontsize=18)
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt


cycles = list(nx.simple_cycles(G))


cycle_nodes = set()
for cycle in cycles:
    cycle_nodes.update(cycle)


C = G.subgraph(cycle_nodes).copy()

pos = nx.spring_layout(C, k=2.5, iterations=200, seed=7)

plt.figure(figsize=(18, 18))

nx.draw_networkx_nodes(
    C, pos,
    node_size=2400,
    node_color="salmon",
    edgecolors="black"
)

nx.draw_networkx_edges(
    C, pos,
    arrowstyle="->",
    arrowsize=22,
    width=2.0
)

nx.draw_networkx_labels(
    C, pos,
    font_size=10,
    font_weight="bold"
)

plt.title("Cycle Dependencies (Recursive / Circular Calls)", fontsize=18)
plt.axis("off")
plt.tight_layout()
plt.show()
```
